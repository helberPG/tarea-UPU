```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE

# Cargar el conjunto de datos
dataframe = pd.read_csv('data.csv', sep=';')

# Separar características (X) y variable objetivo (y)
X = dataframe.drop('Salud_fetal', axis=1)
y = dataframe['Salud_fetal']

# Visualizar la distribución de clases antes del balanceo
plt.figure(figsize=(8, 6))
sns.countplot(x=y)
plt.title('Distribución de clases antes de SMOTE')
plt.xlabel('Salud_fetal')
plt.ylabel('Frecuencia')
plt.show()

# Aplicar SMOTE para balancear las clases
smote = SMOTE(random_state=25)
X_res, y_res = smote.fit_resample(X, y)

# Visualizar la distribución de clases después del balanceo
plt.figure(figsize=(8, 6))
sns.countplot(x=y_res)
plt.title('Distribución de clases después de SMOTE')
plt.xlabel('Salud_fetal')
plt.ylabel('Frecuencia')
plt.show()

# Dividir el conjunto de datos balanceado en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.3, random_state=25)

# Crear y entrenar el clasificador Random Forest
Peso_randomforest = RandomForestClassifier(n_estimators=80, random_state=25)
Peso_randomforest.fit(X_train, y_train)

# Realizar predicciones
predicciones = Peso_randomforest.predict(X_test)

# Generar y visualizar la matriz de confusión
conf_matrix = confusion_matrix(y_test, predicciones)
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap='Blues')
plt.title('Matriz de Confusión')
plt.xlabel('Predicho')
plt.ylabel('Verdadero')
plt.show()

# Imprimir el reporte de clasificación
print(classification_report(y_test, predicciones, target_names=['Normal', 'Sospechoso', 'Patológico']))
```
